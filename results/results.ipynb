```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
t = np.arange(30,140)
```

## Linear

```python
dataset_file_poly = '../results/results_linear.csv'
dataset_linear = pd.read_csv(dataset_file_poly,skiprows=1, names=['predictions'])

predictions = dataset_linear['predictions']
plt.plot(t,predictions)
max_value = predictions.max()
index_of_max = int(predictions.idxmax()) + 30

print(f"Giá trị lớn nhất là {max_value} độ dài data {index_of_max}")
```
