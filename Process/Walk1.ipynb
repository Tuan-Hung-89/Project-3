## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
import os
```

## Import Dataset

```python
dataset_file = '../Data/Acc/Data1/T0_ID00000001_Walk_1.csv' 
dataset_id = os.path.basename(dataset_file).split('_')[1]
dataset = pd.read_csv(dataset_file, skiprows = 2, names=['Ax','Ay','Az'])
N = dataset['Ax'].size 
```

### Visual data

```python
fig, ax = plt.subplots(nrows=1,
                       ncols=3,) 
t_step = 0.01
t = np.arange(0,(N-0.5)*t_step, t_step)

## Ax
Ax_y = dataset['Ax']
ax[0].set(title="Ax")
ax[0].plot(t, Ax_y)

## Ay
Ay_y = dataset['Ay']
ax[1].set(title="Ay")
ax[1].plot(t, Ay_y)

## Az
Az_y = dataset['Az']
ax[2].set(title="Az")
ax[2].plot(t, Az_y)

```

#### Fourier Tranform 

```python
fig, ax = plt.subplots(nrows=1,
                       ncols=3,)

## Ax
Ax = np.fft.fft(Ax_y)
Ax_plot = Ax[0:int(t.size/2+1)]
t_plot = np.linspace(0, 100, Ax_plot.size)
ax[0].plot(t_plot, Ax_plot);

## Ay
Ay = np.fft.fft(Ay_y)
Ay_plot = Ay[0:int(t.size/2+1)]
t_plot = np.linspace(0, 100, Ay_plot.size)
ax[1].plot(t_plot, Ay_plot);

## Az
Az = np.fft.fft(Az_y)
Az_plot = Az[0:int(t.size/2+1)]
t_plot = np.linspace(0, 100, Az_plot.size)
ax[2].plot(t_plot, Az_plot);

frequency = 50  # Tn s cn tm

# Tm v tr ca phn t trong mng t_plot c gi tr bng frequency
indices = np.where(t_plot == frequency)[0]

if len(indices) == 0:
    # Tm v tr ca phn t gn nht vi hoc bng frequency trong mng t_plot
    closest_index = np.argmin(np.abs(t_plot - frequency))
    indices = np.append(indices, closest_index)  # Thm closest_index vo indices
    print("V tr ca phn t gn 50:", indices)

else:
    # In ra v tr ca cc phn t c tn s bng frequency
    print("V tr ca phn t c tn s 50:", indices)

```

#### Clear noise

```python
fig, ax = plt.subplots(nrows=1,
                       ncols=3,)
# Thit k b lc thng thp
cutoff_frequency = indices[0]  # Tn s ct (cutoff frequency)

# p dng b lc
filtered_Ax = Ax_plot[0:cutoff_frequency]
filtered_Ay = Ay_plot[0:cutoff_frequency]
filtered_Az = Az_plot[0:cutoff_frequency]

t_plot2 = np.linspace(0, 50, filtered_Ax.size)

# V biu  tn hiu  lc
ax[0].plot(t_plot2, filtered_Ax)
ax[1].plot(t_plot2, filtered_Ay)
ax[2].plot(t_plot2, filtered_Az)
```
